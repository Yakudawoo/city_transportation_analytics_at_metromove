Count how many routes are underperforming and what percentage of total routes this represents.

```python
# Use route_metrics from Task 16 or recreate it
detailed_metrics = {
    
    
    "Route 101": {
        "daily_trips_count": 24,
        "average_passengers_per_trip": 68,
        "peaks_hours": ["07:00", "08:30", "17:00", "18:30"],
        "percentage_on_time": 94.5  
    },
    "Route 103": {
        "daily_trips_count": 18,        
        "average_passengers_per_trip": 52,
        "peaks_hours": ["08:00", "17:30"],
        "percentage_on_time": 89.2  
    },
    "Route 205": {
        "daily_trips_count": 20,        
        "average_passengers_per_trip": 78,
        "peaks_hours": ["07:30", "18:00"],  
        "percentage_on_time": 91.8  
    }
}

# Identify underperforming routes
underporming_routes = []
for route, metrics in detailed_metrics.items():
    total_passengers = metrics["daily_trips_count"] * metrics["average_passengers_per_trip"]
    if metrics["percentage_on_time"] < 90.0 or total_passengers < 30000:
        reason =[]
        if metrics["percentage_on_time"] < 90.0:
            reason.append(f"On-time performance {metrics['percentage_on_time']}% below 90%")
        if total_passengers < 30000:
            reason.append(f"Total passengers {total_passengers} below 30000")
        underporming_routes.append((route, ", ".join(reason)))

# Report underperforming routes
for route, reasons in underporming_routes:  
    print(f"{route} underperforms due to: {reasons}")
    
print(f"""\nUnderperforming routes: {len(underporming_routes)} routes found.
Percentage: {len(underporming_routes)/len(detailed_metrics)*100}%""" )
```

## Task 19: Create service disruption log

Track service disruptions with detailed records. Each disruption includes: route affected, time, duration (minutes), cause, resolved status.

Create a list where each item is a dictionary representing one disruption:
- Route 101 at 08:15, lasted 25 minutes, cause "mechanical failure", resolved True
- Route 103 at 14:30, lasted 45 minutes, cause "accident", resolved True
- Route 205 at 17:00, lasted 15 minutes, cause "traffic", resolved False
- Route 101 at 18:45, lasted 30 minutes, cause "weather", resolved True

Loop through disruptions. Print each disruption with all details formatted clearly. Calculate total disruption time across all incidents. Identify which route had most disruptions. Flag any unresolved disruptions.

```python
# Create disruption log
disruption_log = [
    {"route": "Route 101", "time": "08:15","duration_minutes": 25, "issue": "mechanical failure", "resolved": True,},
    {"route": "Route 103", "time": "14:30", "duration_minutes": 45, "issue": "accident", "resolved": True},
    {"route": "Route 205", "time": "17:00", "duration_minutes": 15, "issue": "traffic",  "resolved": False},
    {"route": "Route 101", "time ": "18:45","duration_minutes": 30, "issue": "weather", "resolved": True},
]

#initialization
route_disruption_total_time = {}
route_count = {}
# loop through disruptions
for disruption in disruption_log:
    
    # Calculate total disruption time per route #
    route = disruption["route"]
    duration = float(disruption.get("duration_minutes", 0))
    route_disruption_total_time[route] = route_disruption_total_time.get(route, 0) + duration
    
    # Count disruptions per route #
    route_count[disruption['route']] = route_count.get(disruption['route'], 0) + 1
    print(f"Durée totale des perturbations : {route_disruption_total_time} minutes.")

# total disruption time #
for route, total_time in route_disruption_total_time.items():
    print(f"\nTotal disruption time for {route}: {total_time} minutes.")  
    
# most disrupted route #
most_disrupted_route = max(route_count.items(), key=lambda x: x[1])
print(f"\nMost disrupted route: {most_disrupted_route[0]} with {most_disrupted_route[1]} disruptions.")
# most_disrupted_route = max(route_count, key=route_count.get)  # renvoie juste la clé #
# auquel cas on écrirait :
# print(f"\nMost disrupted route: {most_disrupted_route} with {route_count[most_disrupted_route]} disruptions.")

# not resolved yet
print(" Not resolved yet")
for disruption in disruption_log:
    if not disruption.get("resolved", False):   
        print(f"Follow-up needed for disruption on {disruption['route']} on {disruption['time']}.")
```
